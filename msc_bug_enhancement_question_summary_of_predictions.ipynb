Predictions from Llama3 8B for test dataset

```python
predictions_llama3_8b = pd.read_pickle('/Users/isdeniz/Documents/Documents_isdeniz/github/Summary-2406/msc_bug_enhancement_question_test_dataset_llama3_8b_prediction_organized.pkl')
print(predictions_llama3_8b.shape)
```

```python
predictions_llama3_8b.head()
```

```python
actual_labels = predictions_llama3_8b['labels']
predicted_labels = predictions_llama3_8b['llama3_prediction_organized']

confusion_matrix = confusion_matrix(actual_labels, predicted_labels)

class_report = classification_report(actual_labels, predicted_labels, target_names=['bug', 'enhancement', 'question'], digits=3)

conf_matrix_df = pd.DataFrame(confusion_matrix, index=['bug', 'enhancement', 'question'], columns=['bug', 'enhancement', 'question'])
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(conf_matrix_df,
            annot = True,
            fmt = 'd',
            cmap = 'Blues')
plt.xlabel('prediction')
plt.ylabel('actual')
plt.title('Confusion Matrix')
plt.show()
```

Predictions from the nlbse'22 model

```python
predictions = pd.read_csv('/Users/isdeniz/Documents/Documents_isdeniz/github/Summary-0207/msc_bug_enhancement_question/model_nlbse22_prediction_msc_bug_enhancement_question_preprocessing_nlbse22/model_nlbse22_prediction_msc_bug_enhancement_question_preprocessing_nlbse22.csv')
print(predictions.shape)
```

```python
predictions.head()
```

```python
msc_dataset = pd.concat([train, test], ignore_index = True)
msc_dataset.reset_index(drop = True, inplace = True)
msc_dataset.shape[0]
```

```python
msc_dataset.columns
```

```python
predictions.columns
```

```python
summary = pd.DataFrame(columns=['original_issue_title', 
                                'preprocessed_title', 
                                'original_issue_body', 
                                'preprocessed_body',
                                'text', 
                                'labels_name', 
                                'mapLabel', 
                                'nlbse22_prediction',
                                'llama3_8b_prediction',
                                'llama3_70b_prediction'])
```

```python
summary['original_issue_title'] = msc_dataset['original_issue_title']
summary['preprocessed_title'] = msc_dataset['preprocessed_title']
summary['original_issue_body'] = msc_dataset['original_issue_body']
summary['preprocessed_body'] = msc_dataset['preprocessed_body']
summary['text'] = predictions['text']
summary['labels_name'] = msc_dataset['labels_name']
summary['mapLabel'] = msc_dataset['mapLabel']
summary['nlbse22_prediction'] = predictions['prediction']
summary['llama3_8b_prediction'] = msc_dataset['llama3_8b_prediction']
summary['llama3_70b_prediction'] = msc_dataset['llama3_70b_prediction']
```

```python
def get_label_name(text):
  if text == 0:
    return 'bug'
  elif text == 1:
    return 'enhancement'
  elif text == 2:
    return 'question'
```

```python
summary['nlbse22_prediction'] = summary['nlbse22_prediction'].apply(get_label_name)
```

```python
summary.shape[0]
```

```python
actual_labels = summary['mapLabel']
predicted_labels = summary['nlbse22_prediction']
```

```python
label_names = ['bug', 'enhancement', 'question']
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix = confusion_matrix(actual_labels, predicted_labels)
plt.figure(figsize = (8,6))
sns.heatmap(confusion_matrix,
            annot = True,
            fmt = 'd',
            cmap = 'Blues',
            xticklabels = label_names,
            yticklabels = label_names)
plt.xlabel('prediction')
plt.ylabel('actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
summary.head(2)
```

```python
summary.tail(2)
```

```python
summary['nlbse22_check'] = summary['mapLabel'] == summary['nlbse22_prediction']
summary['nlbse22_check'].value_counts()
```

```python
summary['llama3_8b_prediction'].value_counts()
```

```python
summary['train/test'] = 'train'

t = {'bug', 
     'enhancement',
     'question'}

summary.loc[summary['llama3_8b_prediction'].isin(t), 'train/test'] = 'test'

summary['train/test'].value_counts()
```

```python
summary['llama3_8b_check'] = 'NA'

summary.loc[summary['train/test'] == 'test', 'llama3_8b_check'] = summary['mapLabel'] == summary['llama3_8b_prediction']

summary['llama3_8b_check'].value_counts()
```

```python
summary['llama3_70b_check'] = 'NA'

summary.loc[summary['train/test'] == 'test', 'llama3_70b_check'] = summary['mapLabel'] == summary['llama3_70b_prediction']

summary['llama3_70b_check'].value_counts()
```

```python
test_nlbse22 = summary[summary['train/test'] == 'test']
test_nlbse22.reset_index(drop = True)
test_nlbse22['nlbse22_check'].value_counts()
```

```python
test_results = summary[summary['train/test'] == 'test']
test_results = test_results.reset_index(drop = True)
test_results.to_csv('test_results.csv')
```

```python
false_test_results = summary[
    (summary['train/test'] == 'test') & 
    (
        (summary['nlbse22_check'] != True) | 
        (summary['llama3_8b_check'] != True) | 
        (summary['llama3_70b_check'] != True)
    )
    & ~((summary['nlbse22_check'] == True) & (summary['llama3_8b_check'] == True) & (summary['llama3_70b_check'] == True))
]
false_test_results = false_test_results.reset_index(drop = True)
false_test_results.to_csv('false_test_results.csv')
```
